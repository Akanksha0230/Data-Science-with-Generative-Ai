```python
7.#Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type Also display the table attributes and data types of each column the data.
import pandas as pd
import re
data=pd.read_csv('/content/People Data.csv')

# Display the original DataFrame
print("Original DataFrame:")
print(data)

# Function to clean and convert phone numbers to numeric values
def clean_phone(phone):
    if pd.isna(phone):
        return None
    # Remove non-numeric characters and extract digits
    digits = re.sub(r'[^\d]', '', str(phone))
    # Convert to numeric
    try:
        return int(digits)
    except ValueError:
        return None

data['Phone'] = data['Phone'].apply(clean_phone)

# Convert the 'Phone' column to integer dtype
data['Phone'] = data['Phone'].astype('Int64')

# Display the table attributes and data types of each column
print("\nTable Attributes:")
print("Number of rows:", data.shape[0])
print("Number of columns:", data.shape[1])

print("\nData Types of Each Column:")
print(data.dtypes)

# Display the transformed DataFrame
print("\nTransformed DataFrame:")
print(data)
```

```python
14.# Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
#b) Label the x-axis as 'Date'.
#c) Set the title of the plot as 'Temperature and Humidity Over Time'.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset
np.random.seed(0)
dates = pd.date_range(start='2024-01-01', periods=100)
temperature = np.random.randint(10, 30, size=100)
humidity = np.random.randint(30, 70, size=100)

# Create DataFrame
data = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
fig, ax1 = plt.subplots()

# Plot Temperature on left y-axis
ax1.plot(data['Date'], data['Temperature'], color='red')
ax1.set_ylabel('Temperature', color='red')

# Create a second y-axis (right )for Humidity
ax2 = ax1.twinx()
ax2.plot(data['Date'], data['Humidity'], color='blue')
ax2.set_ylabel('Humidity', color='blue')

#b) Label the x-axis as 'Date'.
ax1.set_xlabel('Date')

#c) Set the title of the plot as 'Temperature and Humidity Over Time'.
plt.title('Temperature and Humidity Over Time')

plt.show()
```

```python
15.# Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
#a) Plot a histogram of the data with 30 bins.
#b) Overlay a line plot representing the normal distribution's probability density function (PDF).
#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
#d) Set the title of the plot as 'Histogram with PDF Overlay'.
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array data containing 1000 samples from a normal distribution.
data = np.random.normal(size=1000)

#a) Plot a histogram of the data with 30 bins.
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black', label='Histogram')

#b) Overlay a line plot representing the normal distribution's probability density function (PDF).
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)  # PDF of normal distribution
plt.plot(x, p, 'k', linewidth=2, label='PDF')

#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

#d) Set the title of the plot as 'Histogram with PDF Overlay'.
plt.title('Histogram with PDF Overlay')

# Display legend
plt.legend()

# Show plot
plt.show()
```

```python
17.# Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Define the quadrants
quadrant = np.zeros(100)
quadrant[(x >= 0) & (y >= 0)] = 1
quadrant[(x < 0) & (y >= 0)] = 2
quadrant[(x < 0) & (y < 0)] = 3
quadrant[(x >= 0) & (y < 0)] = 4

# Create a DataFrame for Seaborn
data = {'x': x, 'y': y, 'Quadrant': quadrant}
df = pd.DataFrame(data)

# Create a scatter plot with Seaborn
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set title
plt.title('Quadrant-wise Scatter Plot')

# Show legend
plt.legend(title='Quadrant')

# Display the plot
plt.show()

```

```python
18.#  With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'
from bokeh.plotting import figure, output_notebook, show
import numpy as np

# Output the plot to the notebook
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Plot the sine wave
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)


```

```python
19.# Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np

# Output the plot to the notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))

# Create a Bokeh figure
p = figure(x_range=categories, height=350, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Color bars based on values
colors = ['#FF5733', '#FFC300', '#33FF57', '#338BFF', '#9933FF']
source = ColumnDataSource(data=dict(categories=categories, values=values, color=colors))

# Plot bars
p.vbar(x='categories', top='values', width=0.9, color='color', legend_field="categories", source=source)

# Add hover tooltips
p.add_tools(HoverTool(tooltips=[("Value", "@values")]))

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Remove grid lines
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Set the title
p.title.text = 'Random Categorical Bar Chart'

# Show the plot
show(p)
```

```python
20.# Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.
import plotly.graph_objs as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)
y = np.random.rand(100)

# Create a Plotly trace for the line plot
trace = go.Scatter(x=x, y=y, mode='lines', name='Random Data')

# Create layout
layout = go.Layout(title='Simple Line Plot', xaxis=dict(title='X-axis'), yaxis=dict(title='Y-axis'))

# Create figure
fig = go.Figure(data=[trace], layout=layout)

# Show plot
fig.show()
```
