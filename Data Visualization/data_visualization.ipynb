# **SEABORN ASSIGNMENT:**

```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(42)
x = np.random.rand(100) * 10
y = 2.5 * x + np.random.randn(100) * 5

data = pd.DataFrame({'X Variable': x, 'Y Variable': y})

plt.figure(figsize=(8, 6))
sns.scatterplot(x='X Variable', y='Y Variable', data=data, color='blue', edgecolor='black')
plt.title('Scatter Plot of Synthetic Data', fontsize=14, fontweight='bold')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std Dev=15, 1000 samples
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=30, kde=True, color='purple', edgecolor='black', alpha=0.7)
plt.title('Distribution of a Numerical Variable', fontsize=14, fontweight='bold')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [45, 70, 30, 85, 60]

data = pd.DataFrame({'Category': categories, 'Value': values})

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Value', data=data, palette='viridis')

plt.title('Comparison of Categories Based on Numerical Values', fontsize=14, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
np.random.seed(0)
categories = ['A', 'B', 'C', 'D', 'E']
data = {
    'Category': np.random.choice(categories, size=100),
    'Value': np.random.randn(100) * 10 + 50  # Normally distributed values
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Distribution of Numerical Values Across Categories')
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
from sklearn.datasets import make_regression

np.random.seed(42)
X, _ = make_regression(n_samples=100, n_features=5, noise=0.1, random_state=42)

X[:, 1] = X[:, 0] + np.random.randn(100) * 0.5
X[:, 2] = X[:, 0] + np.random.randn(100) * 0.5
X[:, 3] = X[:, 1] + np.random.randn(100) * 0.5

df = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(5)])

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# **BOKEH ASSIGNMENT:**

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show, output_notebook
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y', width=800, height=400)
p.line(x, y, legend_label="Sine Wave", line_width=2, color="blue")

output_notebook()
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show, output_notebook
import numpy as np
import pandas as pd
np.random.seed(42)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.randint(10, 50, size=n)
colors = np.random.choice(['red', 'green', 'blue', 'purple', 'orange'], size=n)
data = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

p = figure(title="Scatter Plot with Random Sizes and Colors",
           x_axis_label='X', y_axis_label='Y', width=800, height=400)
p.scatter('x', 'y', size='sizes', color='colors', legend_field='colors', source=data, fill_alpha=0.6)

output_notebook()
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
#data_hist = np.random.randn(1000)
#hist, edges = np.histogram(data_hist, bins=30)
from bokeh.plotting import figure, show, output_notebook

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruit', y_axis_label='Count',
           width=800, height=400)

p.vbar(x=fruits, top=counts, width=0.5, color="green", legend_label="Fruit Count")
p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

output_notebook()
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import LinearColorMapper
from bokeh.transform import linear_cmap
import numpy as np

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           width=500, height=500)

color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

p.xaxis.visible = False
p.yaxis.visible = False

output_notebook()

show(p)

```
